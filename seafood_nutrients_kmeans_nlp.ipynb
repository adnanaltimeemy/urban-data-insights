```python
import pandas as pd

# Load the dataset
df = pd.read_csv('global-seafood-nutrient-dataset-raw.csv')

# Display first few rows
print(df.head())

# Display all column names to check for differences
print(df.columns.tolist())
```

```python
# Make column names lowercase, remove spaces and parentheses
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Check again
print(df.columns.tolist())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Example DataFrame
df = pd.DataFrame({
    'age': [25, 32, 47],
    'salary': [50000, 60000, 80000],
    'city': ['NY', 'LA', 'SF']  # non-numeric column
})

# Select only numeric columns
df_numeric = df.select_dtypes(include='number')

# Scale the numeric data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

print(df_scaled)
```

```python
print(df_scaled.shape)
```

```python
from sklearn.cluster import KMeans

# Fit K-Means to your scaled data
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# Add clusters to your original DataFrame
df['Cluster'] = clusters
```
