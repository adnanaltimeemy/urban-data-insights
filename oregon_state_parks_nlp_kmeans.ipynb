```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Example documents
documents = [
    "I love machine learning and AI",
    "Python is great for data science",
    "Artificial intelligence and data analysis",
    "Football and basketball are popular sports",
    "I enjoy watching sports and playing games"
]

# Number of clusters
k = 2

# Step 1: Convert documents into TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Step 2: Fit KMeans
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# Step 3: Get feature names (terms)
terms = vectorizer.get_feature_names_out()

# Step 4: Print top terms per cluster
for i in range(k):
    center_terms = kmeans.cluster_centers_[i]
    top_terms = [terms[ind] for ind in center_terms.argsort()[-10:][::-1]]  # top 10 terms
    print(f"\nCluster {i} Keywords:")
    print(top_terms)
```
