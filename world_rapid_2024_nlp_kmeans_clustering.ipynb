```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Example text data
docs = [
    "I love machine learning",
    "Clustering is fun",
    "Machine learning is amazing",
    "I enjoy coding in Python"
]

# 1. Initialize vectorizer
vectorizer = TfidfVectorizer()

# 2. Fit and transform your text data
X = vectorizer.fit_transform(docs)  # <-- This fits the vocabulary

# 3. Fit KMeans
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# 4. Get top words per cluster
terms = vectorizer.get_feature_names_out()

for i in range(k):
    center = kmeans.cluster_centers_[i]
    top_indices = center.argsort()[-10:][::-1]
    top_terms = [terms[j] for j in top_indices]
    
    print(f"\nCluster {i} top words:")
    print(top_terms)
```
